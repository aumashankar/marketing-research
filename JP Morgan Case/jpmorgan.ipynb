# Data Validation - Total Cells vs Missing %

```python
#Find % of missing data
missing_count = jpmorgan_data.isnull().sum() #number of missing
total_cells = np.product(jpmorgan_data.shape) # number of cells (cols x rows)
total_missing = missing_count.sum()
missing_percent = (total_missing*100)/total_cells

print('Total : ', total_cells)
print('Total missing : ', total_missing)
print('Missing Percentage: ', missing_percent, '%')
```

# Treat null values and missing data

```python
# Handle missing values if necessary

missing_count = jpmorgan_data.isnull().sum() 
print(missing_count)
print(jpmorgan_data.shape)

# drop null values
#features = jpmorgan_data.dropna()

# Calculate the mode for the 'Age' column
age_mode = jpmorgan_data['Age'].mode()[0]

# Replace ages less than 0 with the mode (there are values with -9)
jpmorgan_data['Age'] = jpmorgan_data['Age'].mask(jpmorgan_data['Age'] < 0, age_mode)

# List of columns to be treated with mode
mode_impute_columns = [
    'Q1_Performance_Investments', 'Q1_Fees_Commissions', 'Q1_Depth_Products_Services',
    'Q1_Problem_Resolution', 'Q1_Online_Services', 'Q1_Multiple_Products',
    'Q1_Quality_Advice', 'Q1_Knowledge_Representatives', 'Q1_Understanding_Needs',
    'Q1_Access_Professional_Resources', 'Q1_Provider_Knowledge', 'Q1_Quality_Service',
    'Likelihood_Recommend_Provider', 'Likelihood_Continue_Using_Provider', 
    'Likelihood_Drop_Replace_Provider', 'Advantage_Rating', 'Q6_Overall_Satisfaction', 
    'Q6_Performance_Investments', 'Q6_Fees_Commissions', 'Q6_Depth_Products_Services', 
    'Q6_Problem_Resolution', 'Q6_Online_Services', 'Q6_Multiple_Products',
    'Q6_Quality_Advice', 'Q6_Knowledge_Representatives', 'Q6_Understanding_Needs', 
    'Q6_Access_Professional_Resources', 'Q6_Provider_Knowledge', 'Q6_Quality_Service',
    'Gender', 'Marital_Status','Investment_Decision_Approach','Investment_Decision_Responsibility',
    'Education_Level'
]

# Impute missing values with mode for specified columns
for column in mode_impute_columns:
    most_frequent = jpmorgan_data[column].mode()[0]
    jpmorgan_data[column].fillna(most_frequent, inplace=True)

# Columns to be filled with 0
zero_fill_columns = ['Dependent_Children', 'Other_Dependents']

# Fill missing values with 0
jpmorgan_data[zero_fill_columns] = jpmorgan_data[zero_fill_columns].fillna(0)

# Check if all missing values are treated
missing_count = jpmorgan_data.isnull().sum() 
print(missing_count)
```

```python
# List of factors
factors = ['Q1_Performance_Investments', 'Q1_Fees_Commissions', 'Q1_Depth_Products_Services', 
           'Q1_Problem_Resolution', 'Q1_Online_Services', 'Q1_Multiple_Products', 
           'Q1_Quality_Advice', 'Q1_Knowledge_Representatives', 'Q1_Understanding_Needs', 
           'Q1_Access_Professional_Resources', 'Q1_Provider_Knowledge','Q1_Quality_Service']

# Calculate average importance score for each factor
average_importance = jpmorgan_data[factors].mean().sort_values(ascending=False)

# Display results
display(average_importance)

fig = px.bar(average_importance, labels={'index': 'Factors', 'value': 'Average Importance'},
             title="Average Importance of Each Factor")
fig.show()

# Count the frequency of each reason

# Create an empty DataFrame to store frequency counts
frequency_counts = pd.DataFrame()

# Calculate frequency counts for each factor
for factor in factors:
    frequency_counts[factor] = jpmorgan_data[factor].value_counts()

# Transpose the DataFrame for better readability
frequency_counts = frequency_counts.T

# Display the frequency counts
display(frequency_counts)

# If you want to plot the frequencies, you can use a bar chart for each factor
for factor in factors:
    fig = px.bar(frequency_counts.loc[factor], title=f"Frequency of Scores for {factor}")
    fig.update_layout(xaxis_title="Score", yaxis_title="Frequency")
    fig.show()
```

```python
# 3. Customer segmentation (e.g., using k-means clustering)

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer


# Select relevant features for clustering
features = jpmorgan_data[['Age', 'Gender', 'Q6_Overall_Satisfaction', 'Q6_Quality_Service', 'Q6_Quality_Advice', 'Q1_Performance_Investments']]
# Replace with your actual features

# Convert categorical data to numerical if needed
features['Gender'] = features['Gender'].map({'Male': 1, 'Female': 2})  # Example mapping

# Handle missing values if necessary
#features = features.dropna()

# Impute missing values with mean
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
features = imputer.fit_transform(features)

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)


# Elbow Method
wcss = []
for i in range(1, 11):  # Testing 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plot the results
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()


# Choose the number of clusters based on the elbow plot
k = 11  # For example

kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(scaled_features)

# Add cluster labels to your original DataFrame
jpmorgan_data['Cluster'] = cluster_labels

# Examine each cluster
for i in range(k):
    print(f"Cluster {i}:")
    display(jpmorgan_data[jpmorgan_data['Cluster'] == i].describe())

```
